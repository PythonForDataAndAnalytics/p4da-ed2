Random Numbers and Coins

#### A) Generating random numbers

```python
# import for random number generation and visualization
import numpy.random as npr
import matplotlib.pyplot as plt
```

```python
# the seed allows for reproducibility of pseudorandom numbers
npr.seed(10)
```

```python
# generate a random number from the uniform distribution
npr.rand()
```

```python
# generate another random number
npr.rand()
```

```python
# generate 10 random numbers
n = 10
randomValues = npr.rand(n)
randomValues
```

```python
# check the type of randomValues
type(randomValues)
```

```python
# visualize a sequence of random numbers
plt.scatter(range(len(randomValues)), randomValues)
plt.title(str(n) + ' Random Values')
plt.xlabel('item #')
plt.ylabel('random value')
plt.axhline(y=.5, linestyle='dashed', color='gray')
plt.show()
```

#### B) Simulating coin tosses

```python
# simulate coin tosses
ntosses = 1000
heads = 0
tails = 0
randoms = npr.rand(ntosses)
for random in randoms:
    if random < .5:
        heads += 1
    else:
        tails +=1
print('Heads=', heads, ', Tails=', tails, sep='')
```

```python
# review the distribution of coin tosses
labels = 'Heads', 'Tails'
counts = [heads, tails]
fig, ax = plt.subplots()
ax.pie(counts, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# the randint() function can be used to generate random integers
npr.randint(0,2)
```

```python
# what is the maximum run of heads or tails, in 100 coin tosses?
ntosses = 100
runCount = 0
runSide = None # will track if current run is heads(0) or tails(1); initially neither(None)
runMax = 0
labels = ['H', 'T']
runList = []
for i in range(ntosses):
    side = npr.randint(0,2)
    if runSide == side: # continue the run
        runCount += 1
    else: # reset the run
        if runSide != None: print(' ', end='')
        runList.append(runCount)
        runSide = side
        runCount = 1
    print(labels[side], end='')
    if runCount > runMax:
        runMax = runCount # maintain the max run
print('\nMaximum run is', runMax)
```

```python
# visualize the runs in a sequence of coin tosses
plt.bar(range(len(runList)), runList)
plt.xlabel('run group #')
plt.ylabel('run count')
plt.yticks(range(max(runList)+1))
plt.gca().set_xlim(left=1, right=len(runList))
plt.show()
```
